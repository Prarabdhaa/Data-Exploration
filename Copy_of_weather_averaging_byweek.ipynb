```
import pandas as pd
import seaborn as sns
import matplotlib as plt
```

```
df = pd.read_excel("santa-maria2011-2019.xlsx")
```

```
df.count()
```

```
df['Date'] = pd.to_datetime(df['Date'])
df_new = df.groupby([pd.Grouper(key='Date', freq='W-MON')]).mean()
```

```
df_new.count()
```

```
df_new.reset_index()
```

```
df_new.to_excel('santa-maria2012-19.xlsx')
```
